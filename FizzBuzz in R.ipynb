# FizzBuzz in R

```r
#compact and clean viewing option
x <- 1:100
ifelse(x %% 15 == 0, 'FizzBuzz',
       ifelse(x %% 5 == 0, 'Buzz',
              ifelse(x %% 3 == 0, 'Fizz', x)))
```

```r
for (i in 1:100) {
     if (i %% 3 == 0 & i %% 5 == 0) {print("FizzBuzz")}
     else if (i %% 3 == 0) {print("Fizz")}
     else if (i %% 5 == 0) {print("Buzz")}
     else print(i)}
```
